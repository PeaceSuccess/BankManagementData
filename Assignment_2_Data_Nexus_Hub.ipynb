**CREATING A NEW COLUMN FOR DATE AND YEAR**

```python
#Dictionary mapping for months to numerical values
month_dict = {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6, 'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12}

#Converting month names to numerical representation
bank_marketing_data['month'] = bank_marketing_data['month'].str.lower().map(month_dict)


bank_marketing_data['date']=pd.to_datetime(bank_marketing_data[['day', 'month']].assign(year=2022))

# Extracting year from the combined date column and create a new column tagged 'year'
bank_marketing_data['year'] = bank_marketing_data['date'].dt.year
```

**CREATING A NEW COLUMN FOR CAMPAIGN ID**

```python
# Generate sequential values for the new column with leading zeros making it a four digit number
bank_marketing_data['campaign_id'] = ['{:04d}'.format(i) for i in range(1, len(bank_marketing_data) + 1)]

print(bank_marketing_data.head())
```

```python
# Client table
client_marketing_data = bank_marketing_data[['client_id', 'age', 'job', 'marital', 'education', 'credit_default', 'housing', 'loan']]
# Campaign table
campaign_marketing_data = bank_marketing_data[['campaign_id','client_id', 'duration', 'contact','pdays', 'previous', 'poutcome', 'y', 'date']]
# Economics table
economics_marketing_data = bank_marketing_data[['client_id','emp_var_rate', 'cons_price_idx', 'euribor3m', 'nr_employed']]
```

```python
# Save each DataFrame as a separate CSV file
client_marketing_data.to_csv('client_data.csv', index=False)
campaign_marketing_data.to_csv('campaign_data.csv', index=False)
economics_marketing_data.to_csv('economics_data.csv', index=False)
```
